```python
'''Python Project 4 : Solving Second Order ODE (Simple Pendulum)'''
import numpy as np
from scipy.integrate import odeint
import math
import matplotlib.pyplot as plt
import csv

rng = np.random.default_rng()
```

```python
# Time points
t_span = np.linspace(0,20,200)
```

```python
F = np.zeros_like(t_span)

for n in range(len(F)-1):
    F[n+1] = 0.8 * F[n] +  1 * (rng.random() - 0.5)          
```

```python
plt.plot(F)
```

```python
# Define function
def odefunction(theta,t,b,g,l,m):
  # Define solution array theta
  theta1 = theta[0]
  theta2 = theta[1]
  # define both solution equations
  # theta1: Winkel
  # theta2: Drehrate 
  dtheta1_dt = theta2
  dtheta2_dt = -(b/m)*theta2 - (g/1)*math.sin(theta1) + F[int(t*10)-1]/m 
  # define final solution vector
  dtheta_dt = [dtheta1_dt,dtheta2_dt]
  # Output of Function
  return dtheta_dt

# Input Data
b = 0.5   # Damping Coefficient
g = 9.81   # Acceleration Due to gravity in m/s^2
l = 1      # Length of Pendulum String in meter
m = 1      # Mass in Kg

# Initial condition zero for displacement and 3 for velocity
theta_0 = [0.0,0]

# Solve ODE
result = odeint(odefunction,theta_0,t_span,args=(b,g,l,m))

# Plot Result (Dispalcement and Velocity Change with Time)
plt.plot(t_span,result[:,0],'b-',label=r'$Displacement$')  # Time and displacement plot
plt.plot(t_span,result[:,1],'r--',label=r'$Velocity$')  # Time and velocity plot
plt.ylabel('Displacement and velocity')
plt.xlabel('Time')
plt.legend()
plt.title('Change in dispacement and velocity with repect to time')
plt.show()

# Image generation for animation
#k=1;  # Constant define to get figure for animation
#z=result[:,0]
#for i in z:
#    x0=0
#    y0=0
#    x1=1*math.sin(i) # Position of Ball in x-direction
#    y1=-1*math.cos(i) # Position of Ball in y-direction

    # Define plot file name for generate animation
    #filename = str(k) + '.png'
    #k=k+1
    #print(filename)

    # Plot Animation
    #plt.figure()
    #plt.plot([-1.5,1.5],[0,0],linewidth=4); # Upper line where pendulum is fix
    #plt.plot([x0,x1],[y0,y1],linewidth=2); # Pendulum length string
    #plt.plot(x1,y1,'o',markersize=20); # Ball of Pendulum
    #plt.title('Motion of Pendulum')
    #plt.xlim([-1.5,1.5])
    #plt.ylim([-1.5,1.5])
    #plt.savefig(filename)
  
```
